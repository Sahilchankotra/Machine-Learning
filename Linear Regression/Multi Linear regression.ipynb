```python
#Import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data set
data=pd.read_csv("USA_Housing.csv")
data.head(10)
```

```python
#get the shape of tha data
data.shape
```

```python
# More info about the dataset
data.describe(())
```

```python
# All columns
data.columns
```

```python
#Splitting the dataset
X=data.iloc[:,:-2]
y=data.Price
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
plt.show()
```

```python
# Splitting the data in xtest,xtrain
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#Import regression lib
from sklearn.linear_model import LinearRegression
```

```python
X_train=X_train.values
X_test=X_test.values
y_train=y_train.values
y_test=y_test.values
```

```python
print(X_train)
print(X_train.shape)
```

```python
print(X_test)
print(X_test.shape)
```

```python
print(y_train)
print(y_train.shape)
```

```python
print(y_test)
print(y_test.shape)
```

```python
reg=LinearRegression().fit(X_train, y_train)
reg
```

```python
y_predict= reg.predict(X_test)
y_predict
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(f"Our model accuracy is {r2_score(y_test,y_predict)}")
print(f"Error is: {mean_squared_error(y_test,y_predict)}")
```

```python
# Prediction of one new House 
reg.predict([[6.19075933e+04, 7.01783782e+00, 6.44025575e+00, 3.25000000e+00, 4.38289472e+04]])
```

```python
# Check the best fit line of our mode
plt.scatter(y_test,y_predict)
z=np.polyfit(y_test, y_predict, 1)
p=np.poly1d(z)
plt.plot(y_test,p(y_test), color='red')
plt.show()
```

```python
# Compare the real and prediction value
Predicted_dataframe=pd.DataFrame({"Actual":y_test,"Predicted":y_predict})
Predicted_dataframe.head(15)
```
